# ロボット運動学

## 車輪の段差乗り越え

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import japanize_matplotlib
import math
```

```python
x = []
y = []
y2 = []
for rate in range(1,80):
    #print(rate/10)
    x.append(rate/100)
    shita = math.acos(1-(rate/100))
    y.append(math.sin(shita))
    y2.append(math.tan(shita))

plt.plot(x, y2, lw=3, c="g",label="(1)前輪駆動の前輪乗り上げ時")
plt.plot(x, y2, lw=3,c="g",label="(2)後輪駆動の前輪乗り上げ時")
plt.plot(x, y2,lw=3, c="g",label="(3)前輪駆動の後輪乗り上げ時")
plt.plot(x, y2, lw=3,c="g",label="(4)後輪駆動の後輪乗り上げ時")
plt.title("1周目課題")

plt.savefig('./ロボット設計学徳論1周目課題.png')
plt.legend()
plt.xlabel('段の高さ/車輪半径')
plt.ylabel('必要な摩擦係数')
plt.show()
```
