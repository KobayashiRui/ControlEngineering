## 代数変数の定義
代数を定義する方法は複数存在する  
(`from sympy import *`をしている前提)

代数`x`の定義
1. `x = Symbol('x')`
2. `x = symbols('x')`
3. `var('x')`

## 複数代数の一括定義
代数を複数定義する場合はいっぺんに定義したほうが楽

代数`x`と`y`の定義
1. `x,y = symbols('x y')`
2. `var('x y')`

`symbols`と`var`の引数の形式は同じ

## Sympyによるグラフプロット
SympyのPlotting Moduleをインポートする
→ Matplotlibをバックエンドで使用して2D,3Dグラフを表示する

### 参考
1: [sympy document1.5.1 Plotting](https://docs.sympy.org/latest/modules/plotting.html)

````python
sympy.var('x')
y = sympy.exp(-x**2)
y2 = sympy.exp(-2*x**2)
p1 = sympy.plot(y, (x, -5, 5), ylabel='y', show=False, legend=True)
p2 = sympy.plot(y2, (x, -5, 5), ylabel='y', show=False)
p1.extend(p2)
#p1.save("./test1")
p1.show()
````

````python
sympy.var('x')
y0 = 1/(1+x)
y1 = 1-(x)+(x**2)
y2 = 1-(x)+(x**2)-(x**3)
y3 = 1-(x)+(x**2)-(x**3)+(x**4)
p1 = sympy.plot(y0,y1,y2,y3, (x, -1, 1), ylabel='y', ylim=(-1,10), show=False, legend=True)
p1[0].line_color="black" 
p1[0].label = r"$\dfrac{1}{(1+x)}$"
p1[1].line_color="r"
p1[1].label = r"$1-x+x^2$"
p1[2].line_color="g"
p1[2].label = r"$1-x+x^2-x^3$"
p1[3].line_color="b"
p1[3].label = r"$1-x+x^2-x^3+x^4$"
#p1.save("graph1")
p1.show()
````

````python
sympy.var('x')
y0 = sympy.exp(x)
y1 = 1+x
y2 = 1/(1-x)

p1 = sympy.plot(y0,y1,y2, (x, -1, 1), ylabel='y', ylim=(-1,10), show=False, legend=True)
p1[0].line_color="black" 
p1[0].label = r"$exp(x)$"
p1[1].line_color="r"
p1[1].label = r"$1+x$"
p1[2].line_color="g"
p1[2].label = r"$\dfrac{1}{1-x}$"
p1.save("graph1")
p1.show()
````

````python
sympy.var('x')
y=sympy.exp(x)
y0 = float(y.subs(x,2))
print(float(y0))
for i in range(3,10):
    g = sympy.poly(sympy.series(y,x,0,i))
    display(g)
    result = float(g.subs(x,2).coeffs()[-1])
    diff = (abs((result - y0))/ y0) * 100
    print(result)
    print(diff)
````

##  常微分方程式

# モジュールをインポート
import sympy
from sympy import *

# 数式を MathJax で表示するように設定
#sympy.init_printing()

# y'-y=0

print(sympy.latex(eq_1))

#print(classify_ode(eq_1))

print(sympy.latex(eq_result))

````python
x0 = []
x1 = []
x2 = []
x3 = []
x_map = np.arange(-5,5,0.1)
for x in x_map:
    x0.append(math.e ** x)
    x1.append( 1 + x + ( x**2 )/2 )
    x2.append( ( 1 + x/2 ) / ( 1 - x/2 ) )
    x3.append( 1 / ( 1 - x + (x**2)/2 ) )

plt.plot(x_map, x0, label = "exp(x)")
plt.plot(x_map, x1, label = "[2,0]")
plt.plot(x_map, x2, label = "[1,1]")
plt.plot(x_map, x3, label = "[0,2]")

plt.ylim((-1,10))
plt.xlim((-5,5))
plt.xticks(np.arange(-5,5,1))
plt.grid()
plt.legend()
plt.show()
````
