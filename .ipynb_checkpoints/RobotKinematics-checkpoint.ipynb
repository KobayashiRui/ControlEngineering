## 車輪の段差乗り越え

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import math
```

```python
x = []
y = []
y2 = []
for rate in range(1,60):
    #print(rate/10)
    x.append(rate/100)
    shita = math.acos(1-(rate/100))
    y.append(math.sin(shita))
    y2.append(math.tan(shita))

plt.plot(x, y)
plt.plot(x, y2)
plt.savefig('./test.png')
plt.show()
```
